#Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#using sns 'iris' datset
f_df=sns.load_dataset('iris')
f_df
```

```python
#finding unique number of flower species
f_df['species'].unique()
```

```python
#finding relation is not easy by seeing, we will plot line chart first
plt.plot(f_df.sepal_length,f_df.sepal_width);
```

```python
#we cant find realation with line chart for largre amount of data
#Draw scatter plot from sns
sns.set_style("darkgrid")
sns.scatterplot(x=f_df.sepal_length, y=f_df.sepal_width); # Pass x and y as keyword arguments
```

```python
#Adding hues to distinguish flowers
sns.scatterplot(x=f_df.sepal_length, y=f_df.sepal_width, hue=f_df.species, s=100);
```

```python
#Customizing the plots
plt.figure(figsize=(15, 5))
plt.title('Iris Sepals')

sns.scatterplot(x=f_df.sepal_length, y=f_df.sepal_width, hue=f_df.species, s=100);
```

## Plotting using Pandas Dataframe

```python
plt.figure(figsize=(15, 5))
plt.title('Iris Sepals')

sns.scatterplot(x='sepal_length',
                y='sepal_width',
                hue="species",
                s=100,
                data=f_df);
```
