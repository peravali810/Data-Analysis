#Line chart Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
apple_yeild=[0.43,0.76,0.56,0.98,0.83]
```

```python
plt.plot(apple_yeild)
```

```python
plt.plot(apple_yeild);
```

```python
#customizing the x-axis
years=['2018','2019','2020','2021','2022']
plt.plot(years,apple_yeild);
```

```python
#adding axis labels
plt.plot(years,apple_yeild)
plt.xlabel('Year')
plt.ylabel('Yeild');
```

```python
#Plotting Multiple Lines
year=['2018','2019','2020','2021','2022','2023','2024','2025','2026']
apple=[100,80,90,110,120,130,120,110,100]
banana=[20,60,70,90,80,70,60,50,40]
plt.plot(year,apple)
plt.plot(year,banana)
plt.xlabel('Year')
plt.ylabel('Yeild');
```

```python
#Adding legend and title
plt.plot(year,apple)
plt.plot(year,banana)
plt.xlabel('Year')
plt.ylabel('Yeild')
plt.legend(['Apple','Banana']);
plt.title('Yeild of Apple and Banana in India');
```

```python
#Adding line markers
plt.plot(year,apple,marker='o')
plt.plot(year,banana,marker='*')
plt.xlabel('Year')
plt.ylabel('Yeild')
plt.legend(['Apple','Banana']);
plt.title('Yeild of Apple and Banana in India');
```

```python
#Styling line and colors
plt.plot(year,apple,marker='o',color='red',ls='--')
plt.plot(year,banana,marker='*',color='green',ls=':')
plt.xlabel('Year')
plt.ylabel('Yeild')
plt.legend(['Apple','Banana']);
```

```python
#Changing figure size
plt.figure(figsize=(10,5))
plt.plot(year,apple,marker='o',color='red',ls='--')
plt.plot(year,banana,marker='*',color='green',ls=':')
plt.xlabel('Year')
plt.ylabel('Yeild');
```
