#Heat Map

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set_style('darkgrid')
```

```python
f_df=sns.load_dataset('flights')
f_df.head()
```

```python
plt.plot(f_df.passengers);
```

```python
#using pivot for making matrix
smalldf=sns.load_dataset('flights').pivot(columns="year",index="month",values="passengers")
smalldf.head()
```

```python
plt.title("Passesngers Count")
sns.heatmap(smalldf);
```

```python
plt.title("Passesngers Count")
sns.heatmap(smalldf, fmt="d", annot=True, cmap="YlGnBu");
```
