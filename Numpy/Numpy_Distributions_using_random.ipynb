# Numpy: Distibutions using Random function


```python
import numpy as np
from numpy import random
```

```python
# random integer generator
x = random.randint(100)
y=random.randint(100, size=(3,5)) #with size of array

print(x)
print(y)
```

```python
#random float generator

x = random.rand()

print(x)
```

```python
#random from choice
x = random.choice([3, 5, 7, 9])
y = random.choice([3, 5, 7, 9], size=(3, 5)) # with size

print(x)
print(y)
```

## Data/probability Distribution

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(20))
y = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5)) #with array size

print(x)
print(y)
```

## Random Permutations/Shuffle

```python
arr = np.array([1, 2, 3, 4, 5])
random.shuffle(arr)#original changed

arr1 = np.array([1, 2, 3, 4, 5])

print(arr)
print(random.permutation(arr1))#original unchanged
print(arr1)
```

## Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

#plotting distplot with histogram
sns.distplot([0, 1, 2, 3, 4, 5])
plt.show()
```

```python
#plotting using histplot and kdeplot
sns.histplot([0, 1, 2, 3, 4, 5])
plt.show()

sns.kdeplot([0, 1, 2, 3, 4, 5])
plt.show()
```

## Normal/ Guassian Distribution

```python
x = random.normal(size=(2, 3))
y = random.normal(loc=1, scale=2, size=(2, 3)) #with mean and standard deviation

print(x)

print(y)
```

```python
#visualisation of ND
sns.displot(random.normal(size=1000), kind='kde')
plt.show()
```

## Binomial Distribution

```python
x = random.binomial(n=10, p=0.5, size=10)
print(x)
```

```python
#Visualization of ND and BD
sns.kdeplot(random.normal(loc=50, scale=5, size=1000), label='Normal Distribution')
sns.kdeplot(random.binomial(n=100, p=0.5, size=1000), label='Binomial Distribution')
plt.legend()
plt.show()
```

##Poisson Distribution

```python
x = random.poisson(lam=1, size=[2,4])

print(x)
```

```python
#Visualization of Poisson distribution
sns.kdeplot(random.poisson(lam=2, size=1000), label='Poisson Distribution')
plt.show()
```

## Uniform Distribution

```python
x = random.uniform(size=(2, 3))
print(x)

sns.kdeplot(random.uniform(size=(2,2)), label='Uniform Distribution')
plt.show()
```

## Logistic Distribution

```python
x = random.logistic(loc=1, scale=2, size=(2, 3))
print(x)

#visualization
sns.kdeplot(random.normal(scale=2, size=1000), label='normal')
sns.kdeplot(random.logistic(random.logistic(size=1000)), label='logistic')
plt.legend()
plt.show()
```
