# Cleaning data with pandas

```python
import pandas as pd
```

## Empty Cells

```python
#Removing rows wiht empty cells

df = pd.read_csv('data.csv')
print(df.to_string())
```

```python
new_df = df.dropna()
print(new_df.to_string())
```

```python
#chaning the original file
df.dropna(inplace = True)

print(df.to_string())
```

```python
#replacing null values with 20
df = pd.read_csv('data.csv')
df.fillna(20, inplace = True)
print(df.to_string())
```

```python
##Replacing using Mean, Median and Mode
df = pd.read_csv('data.csv')
x = df["Calories"].mean()
df["Calories"].fillna(x, inplace = True)
print(df.to_string())
```

```python
#Replacin wrong format
import pandas as pd

df = pd.read_csv('data.csv')
df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True, errors='coerce')

print(df.to_string())
```

```python
#Replacing values
df.loc[7, 'Duration'] = 45
print(df.to_string())
```
